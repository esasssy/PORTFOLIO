```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from pandas.plotting import table

# Step 1: Create the DataFrame for the table
data = {
    "Step": [
        "Customer Accesses Pre-Check Page",
        "Customer Provides Basic Details",
        "Soft Credit Inquiry",
        "Eligibility Calculation",
        "Display Results to Customer"
    ],
    "Description": [
        "Customer navigates to the loan eligibility pre-check feature on the mobile app.",
        "Customer enters basic information (e.g., income, employment status).",
        "The system performs a soft credit inquiry using a third-party credit bureau API.",
        "The system calculates the eligibility based on credit score, income, and other provided details.",
        "The system displays the eligibility result to the customer, with personalized loan offers if eligible."
    ],
    "Data Inputs": [
        "Customer ID",
        "Income details, Employment status",
        "Customer ID, Credit bureau API",
        "Credit score, Income, Employment status",
        "Eligibility result, Loan products available"
    ],
    "Outputs": [
        "Pre-check interface is loaded.",
        "Input data is validated, Proceed to next step.",
        "Credit score is retrieved without affecting the score.",
        "Eligibility result is generated (Eligible/Not Eligible)",
        "Results displayed in a user-friendly format."
    ],
    "Assumptions/Dependencies": [
        "Customer is logged in.",
        "Customer provides accurate information.",
        "Integration with the credit bureau API is functional.",
        "Eligibility criteria are predefined and up-to-date.",
        "Customer understands the result and next steps."
    ]
}

# Step 2: Create the table using pandas and matplotlib
df = pd.DataFrame(data)

# Create a plot for the table
fig, ax = plt.subplots(figsize=(16, 10))  # Increase figure size
ax.axis('off')  # Hide the axis

# Add the table to the plot
tbl = table(ax, df, loc='center', cellLoc='center', colWidths=[0.2]*len(df.columns))

# Customize the table appearance
tbl.auto_set_font_size(False)
tbl.set_fontsize(12)  # Larger font size for readability
tbl.scale(1.5, 1.5)  # Adjust table size
tbl.auto_set_column_width([0, 1, 2, 3, 4])

# Add borders and background color
for (i, j), cell in tbl.get_celld().items():
    if i == 0:
        cell.set_fontsize(14)  # Header font size
        cell.set_text_props(weight='bold')
    else:
        cell.set_edgecolor('black')
        cell.set_linewidth(1)
        cell.set_facecolor('#f5f5f5')  # Light grey background

# Save the table as an image
plt.savefig("Functional_Specification_Table.png", bbox_inches='tight', dpi=300)
plt.show()

# Step 3: Create the flowchart using matplotlib
fig, ax = plt.subplots(figsize=(12, 10))

# Define the positions of the boxes
box_positions = {
    "Start": (0.5, 0.9),
    "Customer Accesses Pre-Check Page": (0.5, 0.75),
    "Customer Provides Basic Details": (0.5, 0.6),
    "Soft Credit Inquiry": (0.5, 0.45),
    "Eligibility Calculation": (0.5, 0.3),
    "Display Results to Customer": (0.5, 0.15),
    "End": (0.5, 0.0)
}

# Draw the boxes
for key, (x, y) in box_positions.items():
    rect = patches.FancyBboxPatch(
        (x - 0.3, y - 0.1), 0.6, 0.2,
        boxstyle="round,pad=0.1", 
        edgecolor='black', facecolor='pink'
    )
    ax.add_patch(rect)
    plt.text(x, y, key, ha='center', va='center', fontsize=14, weight='bold')

# Draw the arrows
for (start, end) in zip(list(box_positions.keys())[:-1], list(box_positions.keys())[1:]):
    start_x, start_y = box_positions[start]
    end_x, end_y = box_positions[end]
    plt.annotate(
        "", xy=(end_x, end_y + 0.1), xytext=(start_x, start_y - 0.1),
        arrowprops=dict(arrowstyle="->", color='black', lw=2)
    )

# Set the plot limits and hide axes
ax.set_xlim(0, 1)
ax.set_ylim(-0.1, 1.1)
ax.axis('off')  # Hide the axes

# Save the flowchart as an image
plt.savefig("Loan_Eligibility_Flowchart.png", bbox_inches='tight', dpi=300)
plt.show()

print("Flowchart saved as 'Loan_Eligibility_Flowchart.png' and table saved as 'Functional_Specification_Table.png'.")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_flowchart(positions, filename, box_color, arrow_color):
    """
    Draws a flowchart with specified positions and saves it as an image file.
    
    Parameters:
    positions (dict): Dictionary with step names and their positions.
    filename (str): The filename to save the flowchart image.
    box_color (str): Color for the boxes.
    arrow_color (str): Color for the arrows.
    """
    fig, ax = plt.subplots(figsize=(12, 10))

    # Draw the boxes
    for key, (x, y) in positions.items():
        rect = patches.FancyBboxPatch(
            (x - 0.3, y - 0.1), 0.6, 0.2,
            boxstyle="round,pad=0.1", 
            edgecolor='black', facecolor=box_color
        )
        ax.add_patch(rect)
        plt.text(x, y, key, ha='center', va='center', fontsize=14, weight='bold')

    # Draw the arrows
    keys = list(positions.keys())
    for i in range(len(keys) - 1):
        start = keys[i]
        end = keys[i + 1]
        start_x, start_y = positions[start]
        end_x, end_y = positions[end]
        plt.annotate(
            "", xy=(end_x, end_y + 0.1), xytext=(start_x, start_y - 0.1),
            arrowprops=dict(arrowstyle="->", color=arrow_color, lw=2)
        )

    # Set the plot limits and hide axes
    ax.set_xlim(0, 1)
    ax.set_ylim(-0.2, 1.1)
    ax.axis('off')  # Hide the axes

    # Save the flowchart as an image
    plt.savefig(filename, bbox_inches='tight', dpi=300)
    plt.show()

# Define the positions of the boxes for system integration
system_integration_positions = {
    "Start": (0.5, 0.9),
    "Define Integration Requirements": (0.5, 0.75),
    "Develop Integration Interfaces": (0.5, 0.6),
    "Data Mapping and Transformation": (0.5, 0.45),
    "Test Integration": (0.5, 0.3),
    "Deploy Integration": (0.5, 0.15),
    "Monitor and Maintain Integration": (0.5, 0.0),
    "End": (0.5, -0.15)
}

# Draw and save the flowchart for system integration
draw_flowchart(system_integration_positions, "System_Integration_Flowchart.png", 'lavender', 'black')

# Define the positions of the boxes for regulatory compliance
compliance_positions = {
    "Start": (0.5, 0.9),
    "Implement Responsible Lending Practices": (0.5, 0.75),
    "Affordability Assessment": (0.5, 0.6),
    "Transparent Loan Terms": (0.5, 0.45),
    "Data Privacy and Security Measures": (0.5, 0.3),
    "Data Encryption": (0.5, 0.15),
    "Customer Consent": (0.5, 0.0),
    "End": (0.5, -0.15)
}

# Draw and save the flowchart for regulatory compliance
draw_flowchart(compliance_positions, "Regulatory_Compliance_Flowchart.png", 'pink', 'black')
```
