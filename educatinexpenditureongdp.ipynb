```python
import pandas as pd
import statsmodels.api as sm

# Load data from Excel file
file_path = "C:\\Users\ACER\OneDrive\Desktop\PD\MSc ECO\DISSERTATION\\us.xlsx"  
df = pd.read_excel(file_path, sheet_name="Sheet1")  # Update sheet_name if necessary

# Ensure the dataframe has the necessary columns
# Columns should be 'Education_Expenditure_%_of_GDP' and 'GDP_Growth_Rate'
assert 'edu' in df.columns, "Column 'edu' not found in the data."
assert 'gdp' in df.columns, "Column 'gdp"

# Independent variable (X) - Education Expenditure as a percentage of GDP
X = df['edu']

# Dependent variable (y) - GDP Growth Rate
y = df['gdp']

# Add a constant to the independent variable (X) for the intercept term in the regression model
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())
```
